## Q.1

```python
import seaborn as sb
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),linestyle="-.",color="k")
plt.xlabel("X-axis",color="g",size=20)
plt.ylabel("Y-axis",color="g",size=20)
plt.title("Sine Wave ",color="red",size=20)
plt.show()
```

## Q.3 
- Plot a line curve with list=[2,3,4,5], use linestyle and color together for the graph and marker style as "Rotated Hexagon Marker"

```python
a=[2,3,4,5]
a=np.array(a)
plt.plot(a,a,"g-.",marker="H",linewidth=2)
plt.show()
```

```python
data={"Store":["Delhi","Mumbai","Chennai","Kolkata","Hyderabad"],"Clothing":[82,78,48,46,39],"Equipment":[48,31,51,29,18],"Accessories":[15,20,14,13,9]}
shw=pd.DataFrame(data)
plt.bar(shw.Store,shw.Clothing,color="gold",label="Clothing",width=0.8,bottom=shw.Equipment+shw.Accessories)
plt.bar(shw.Store,shw.Equipment,color="g",label="Equipment",width=0.8,bottom=shw.Accessories)
plt.bar(shw.Store,shw.Accessories,color="silver",label="Accessories",width=0.8)
plt.legend(loc=1)
plt.xticks(rotation=90)
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
data=sb.load_dataset("titanic")
data.head()
```

```python
sb.catplot(x="survived", y="pclass", hue="class", kind="bar", data=data, palette="pastel")
plt.show()
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
data=sb.load_dataset("tips")
data.head()
```

```python
sb.distplot(data["total_bill"],vertical=True,color="g",kde=False)
plt.xlabel("Total bill vary to x-axis")
plt.title("Total bill Graph")
plt.show()
```

- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow
